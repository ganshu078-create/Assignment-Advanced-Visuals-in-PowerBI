**Q1. What is the difference between a Basic Filter, Advanced Filter, and a Slicer in Power BI?**

- Basic Filters are simple value-based filters, Advanced Filters allow condition-based filtering, and Slicers provide an interactive way for users to filter report data.

| Filter Type     | Description                                                                                |
| --------------- | ------------------------------------------------------------------------------------------ |
| Basic Filter    | Used to select specific values from a list. Example: Display sales only for the year 2023. |
| Advanced Filter | Used for applying conditions such as greater than, less than, contains, starts with, etc.  |
| Slicer          | A visual filter placed on the report page that allows users to interactively filter data.  |

**Q4. What is a Visual-Level Filter, and how does it differ from Page-Level and Report-Level Filters?**

- A Visual-Level Filter may show only Laptop sales in one chart, while other charts remain unchanged.

| Filter Type         | Scope                                      |
| ------------------- | ------------------------------------------ |
| Visual-Level Filter | Affects only one visual.                   |
| Page-Level Filter   | Affects all visuals on a single page.      |
| Report-Level Filter | Affects every visual in the entire report. |

**Q6. What is the role of Drill-Down in Power BI? How does it improve data exploration?**

- Answer:

Drill-Down allows users to move from a higher-level summary to a more detailed level within a hierarchy.

Benefits:

Provides deeper insights.

Helps identify trends and patterns.

Reduces report clutter.

Makes data analysis more interactive.

- Example:

Country → Region → Product Category

Users can start from country-level sales and drill down to individual product categories.

**Q13. What is the “Keep All Filters” option in Drill-Through? How does it affect user interaction?**

Answer:

The Keep All Filters option passes all currently applied filters to the Drill-Through page.

Benefits:

- Maintains analysis context.
- Saves time.
- Improves user experience.

Example:

If a user filters:

Country = India

Region = North

and then drills through to Laptop details, those filters remain applied automatically.
